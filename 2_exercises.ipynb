## Q6





```python
import scipy.stats as stats

s1_sq = 3.6**2
s2_sq = 2.1**2
n1 = 10
n2 = 16
alpha = 0.05

df1 = n1 - 1
df2 = n2 - 1

F_lower = stats.f.ppf(alpha/2, df1, df2)
F_upper = stats.f.ppf(1 - alpha/2, df1, df2)

ratio = s1_sq / s2_sq

ci_lower = ratio / F_upper
ci_upper = ratio / F_lower

print(ci_lower, ci_upper)
```
