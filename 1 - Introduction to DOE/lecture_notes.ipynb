## What is DOE & why it matters

- **Experiment:** a test or series of tests to learn about a system or process. Used widely in engineering & physical sciences (process characterization/optimization, material properties, product design, tolerance setting).
- **Ultimate goals:**
  - Shorten time to design/develop products & processes.
  - Improve performance of existing processes.
  - Improve product reliability & performance.
  - Achieve robustness to variation (products/processes).

---

## A (very) abbreviated history of DOE

- **Agricultural origins (1918–1940s)** — R. A. Fisher & co.; factorial designs, ANOVA; huge impact on agriculture.
- **First industrial era (1951–late 1970s)** — Box & Wilson; response surface methods; emphasis on *immediacy* and *sequential* experimentation; strong in chemical/process industries.
- **Second industrial era (late 1970s–1990)** — Quality movements; Taguchi methods; robust parameter design and process robustness.
- **Modern era (≈1990–)** — Widespread industrial adoption and software support.

### Notable figures
- **William S. Gosset (“Student”)** — Emphasized both improving average yield and breeding for **insensitivity to soil/climate** (robustness).
- **R. A. Fisher** — Foundational work in design, randomization, ANOVA; introduced factorial concepts.
- **George E. P. Box** — Co-developed response surfaces; championed sequential experimentation.

## Core principles of good experiments

- **Randomization**  
  Run trials in a **random order** to balance out unmeasured “lurking” variables.

- **Replication**  
  Provides precision (reduces standard errors) and enables estimation of **pure error/background noise**.  
  Distinguish **replicates** (independent runs) vs. **repeat measurements** (multiple reads on the same run).

- **Blocking**  
  Group similar experimental units to handle **nuisance factors** (known sources of variation you’re not studying).

---

## Strategy of experimentation

- **“Best-guess” trials:** common and sometimes surprisingly successful, but risk bias and inefficiency.
- **One-factor-at-a-time (OFAT):** often aligned with naïve “scientific method,” but **inefficient** and **wrecked by interactions**.
- **Statistically designed experiments:** Based on **factorials**; capture interactions; efficient use of data.

---

## Factorial designs (the workhorse)

- **Definition:** test **all combinations** of factor levels. Reveals **main effects** and **interactions**.
- **Scaling up:** With many factors, full factorials can be large → use **fractional factorials** to screen efficiently.
- **Illustrative “golf experiment”:** (factors like driver, ball, walking vs. riding, beverage, time, weather, spikes, …) emphasizes multi-factor thinking.

---

## Planning → Conducting → Analyzing (the 7-step loop)

1. **Recognize & state the problem** — Are you characterizing, optimizing, or seeking robustness?
2. **Select response variable(s)** — mean, variance, or multiple responses as needed.
3. **Choose factors, levels, ranges**
   - Include **design factors**, **held-constant factors**, and **allowed-to-vary** factors.
   - Classify **nuisance factors**: controllable, uncontrollable, or noise.
4. **Choose the design** — sample size, run order, and structure (e.g., blocks, factorial vs. fractional).
5. **Conduct the experiment** — monitor/mitigate **experimental error** while running.
6. **Statistical analysis** — use software and **graphical methods**; perform **residual analysis** & **model adequacy checks**.
7. **Conclusions & recommendations** — plan **follow-up/confirmation runs**; proceed **sequentially/iteratively**.

### Process tips
- Involve **statistical thinking early**.
- **Subject-matter knowledge** is crucial.
- **Pre-experimental planning (Steps 1–3) is vital**.
- Think **sequentially**; improve via **confirmation tests**.

## Quick glossary (from the chapter)

- **Factor:** controllable input you vary (e.g., temperature). **Level:** specific setting of a factor.
- **Response:** output you measure (e.g., yield, time, variance).
- **Interaction:** when the effect of one factor depends on the level of another; OFAT easily misses this.
- **Nuisance factor:** source of variation not of primary interest; handle via blocking or randomization.
- **Robustness:** performance insensitive to noise (echoing Gosset/Taguchi motivation).
