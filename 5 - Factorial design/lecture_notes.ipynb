# Chapter 5 – Factorial Designs

## 1. Introduction
- **Factorial designs** allow us to study the effect of **two or more factors simultaneously**.
- General principles:
  - Efficient compared to one-factor-at-a-time.
  - Allow estimation of **main effects** and **interaction effects**.
- Reference: Montgomery (2012), Ch. 5.

---

## 2. Factor Effects
- **Definition:** The effect of a factor is the change in mean response when the factor goes from low (−1) to high (+1).

If $y_{A+}$ and $y_{A-}$ are the averages at high/low levels:
$$
\text{Effect of A} = \frac{y_{A+} - y_{A-}}{2}
$$

Example:
- $A = \frac{40 + 52}{2} - \frac{20 + 30}{2} = 21$
- $B = \frac{30 + 52}{2} - \frac{20 + 40}{2} = 11$






**Interaction effect**: $AB = \frac{52 + 20}{2} - \frac{30 + 40}{2} = 1$

---

## 3. Regression Model & Response Surface
Coded variables: $x_i \in \{-1, +1\}$

General regression model for two factors:
$$
\hat{y} = b_0 + b_1x_1 + b_2x_2 + b_{12}x_1x_2
$$

- $b_0 =$ Grand mean
- $b_i = \tfrac{1}{2} \cdot (\text{Effect of factor i})$
- $b_{12} = \tfrac{1}{2} \cdot (\text{Interaction Effect})$



Example:
$$
\hat{y} = 35.5 + 10.5x_1 + 5.5x_2 - 0.5x_1x_2
$$

## 4. Interactions
- Interaction exists if the effect of one factor **depends on the level of another factor**.
- Response surface shows curvature if interaction is present.

Model with interaction:
$$
\hat{y} = b_0 + b_1x_1 + b_2x_2 + b_{12}x_1x_2
$$

---

## 5. General Two-Factor Factorial Experiment
- Factors: $A$ (levels = $a$), $B$ (levels = $b$), with $n$ replicates.
- **Completely randomized design**.
- **Statistical effects model**:
$$
y_{ijk} = \mu + \tau_i + \beta_j + (\tau\beta)_{ij} + \epsilon_{ijk}
$$
where:
- $\mu$: overall mean
- $\tau_i$: effect of level $i$ of factor $A$
- $\beta_j$: effect of level $j$ of factor $B$
- $(\tau\beta)_{ij}$: interaction effect
- $\epsilon_{ijk} \sim N(0, \sigma^2)$

## 6. ANOVA for Factorials (Fixed Effects Case)

### Sums of Squares
$$
SS_T = \sum\sum\sum (y_{ijk} - \bar{y}_{...})^2
$$
$$
SS_A = bn \sum_{i=1}^a (\bar{y}_{i..} - \bar{y}_{...})^2
$$
$$
SS_B = an \sum_{j=1}^b (\bar{y}_{.j.} - \bar{y}_{...})^2
$$
$$
SS_{AB} = n \sum_i \sum_j (\bar{y}_{ij.} - \bar{y}_{i..} - \bar{y}_{.j.} + \bar{y}_{...})^2
$$
$$
SS_E = SS_T - SS_A - SS_B - SS_{AB}
$$

### Degrees of Freedom
- $df_A = a-1$
- $df_B = b-1$
- $df_{AB} = (a-1)(b-1)$
- $df_E = ab(n-1)$
- $df_T = abn - 1$

### ANOVA Table


| Source       | df              | SS             | MS = SS/df   | F ratio             |
|--------------|----------------|----------------|--------------|---------------------|
| Factor A     | $a-1$          | $SS_A$         | $MS_A$       | $F_A = MS_A/MS_E$   |
| Factor B     | $b-1$          | $SS_B$         | $MS_B$       | $F_B = MS_B/MS_E$   |
| Interaction  | $(a-1)(b-1)$   | $SS_{AB}$      | $MS_{AB}$    | $F_{AB}=MS_{AB}/MS_E$|
| Error        | $ab(n-1)$      | $SS_E$         | $MS_E$       |                     |
| Total        | $abn-1$        | $SS_T$         |              |                     |

## 7. Quantitative & Qualitative Factors
- **ANOVA** treats all factors as qualitative.
- In practice, many factors are **quantitative** (e.g., temperature, pressure).
- Use **regression models** to fit **response curves/surfaces**:
  - Linear, quadratic, cubic terms in quantitative variables.
  - Example terms: $B$, $B^2$, $AB$, $AB^2$, $B^3$.

This allows prediction and optimization.




---

## 8. Hierarchy Principle
- If higher-order terms (e.g., interaction, quadratic) are included, their lower-order terms should usually also be included.
- Example:
  - If $AB^2$ is significant, the model should also include $A$, $B$, $AB$, $B^2$.

---

## 9. Factorials with More Than Two Factors
- Generalize to 3 or more factors.
- All treatment combinations ($a \times b \times c \dots$) are tested.
- ANOVA identity extends:
$$
SS_T = SS_A + SS_B + SS_C + SS_{AB} + SS_{AC} + SS_{BC} + SS_{ABC} + \dots + SS_E
$$

---

## 10. Key Takeaways
- Factorial designs study multiple factors simultaneously.
- Estimate **main effects** and **interactions**.
- Use **ANOVA** for hypothesis testing.
- Use **regression models** for prediction/response surfaces.
- Be mindful of **hierarchy principle** when building models.
- Easily extended to more than 2 factors.
