## ðŸ”¹ Blocking Principle
- **Purpose**: Reduce the effect of **nuisance factors** (unwanted variability) on the response.
- **Definition**: A nuisance factor is something that influences the response but is **not of primary interest**.
- **Typical nuisance factors**: 
  - Batches of raw material
  - Operators
  - Test equipment
  - Time (shifts, days, etc.)
- **Handling strategies**:
  - If nuisance factor is **known & controllable** â†’ use **blocking**.
  - If **known but uncontrollable** â†’ sometimes use **ANCOVA**.
  - If **unknown & uncontrollable** â†’ rely on **randomization**.

## ðŸ”¹ Randomized Complete Block Design (RCBD)
- **Idea**: Divide experimental units into *blocks* (groups that are internally homogeneous).  
- **Design**: Each block contains all treatments (a complete replicate).
- **Randomization**: Treatments are randomized *within each block*.
- **Effect**: Removes variation due to blocks from the error term.

### Statistical Model
For $a$ treatments and $b$ blocks:

$$
y_{ij} = \mu + \tau_i + \beta_j + \epsilon_{ij}
$$

- $y_{ij}$ = response from treatment $i$ in block $j$  
- $\mu$ = overall mean  
- $\tau_i$ = effect of treatment $i$  
- $\beta_j$ = effect of block $j$ (nuisance)  
- $\epsilon_{ij}$ = random error  

### Hypotheses
- $H_0$: $\tau_1 = \tau_2 = \cdots = \tau_a = 0$ (no treatment effect)  
- $H_A$: At least one $\tau_i \neq 0$  

### ANOVA Partitioning
$$
SS_T = SS_{\text{Trt}} + SS_{\text{Blocks}} + SS_E
$$

Degrees of freedom:  
- Treatments: $a-1$  
- Blocks: $b-1$  
- Error: $(a-1)(b-1)$  
- Total: $ab-1$

Test statistic:
$$
F = \frac{MS_{\text{Trt}}}{MS_E}
$$

## ðŸ”¹ Examples
- **Hardness testing**: Blocks = coupons of different hardness, treatments = different tips.  
- **Vascular graft**: Blocks = resin batches, treatments = extrusion pressures.

## ðŸ”¹ Latin Square Design
- Used when **two nuisance factors** must be controlled simultaneously.  
- Structure: $p \times p$ square, where each treatment appears once in each row and column.  
- **Assumption**: No interaction between treatments and nuisance factors.  

### Model
$$
y_{ijk} = \mu + \alpha_i + \tau_j + \beta_k + \epsilon_{ijk}
$$

- Row effect = $\alpha_i$ (nuisance 1)  
- Column effect = $\beta_k$ (nuisance 2)  
- Treatment effect = $\tau_j$  

### Use case
- Example: Rocket propellant testing (treatments = propellant types, rows = batches, columns = test days).
