# Mean and Variance for Discrete Distributions

---

## 1. Mean (Expectation)

- **Definition:**
  For a discrete random variable $X$ with probability mass function $f(x) = P(X=x)$:

  $$
  \mu = \mathbb{E}[X] = \sum_x x \, f(x)
  $$

- **Interpretation:**  
  The mean is the long-run average value of $X$ if we repeat the random experiment many times.

---

## 2. Variance

- **Definition:**
  $$
  \sigma^2 = \text{Var}(X) = \mathbb{E}[(X-\mu)^2] = \sum_x (x-\mu)^2 f(x)
  $$

- **Shortcut formula:**
  $$
  \text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2
  $$
  where
  $$
  \mathbb{E}[X^2] = \sum_x x^2 f(x)
  $$

- **Standard deviation:**
  $$
  \sigma = \sqrt{\text{Var}(X)}
  $$

---

## 3. Example: Fair Dice ðŸŽ²

- Random variable $X$ = outcome of one throw ($1,2,3,4,5,6$), each with probability $f(x) = \tfrac{1}{6}$.

- **Mean:**
  $$
  \mu = \sum_{x=1}^6 x \cdot \tfrac{1}{6} = \tfrac{1+2+3+4+5+6}{6} = 3.5
  $$

- **Variance:**
  $$
  \sigma^2 = \sum_{x=1}^6 (x-3.5)^2 \cdot \tfrac{1}{6} \approx 2.92
  $$

- **Standard deviation:**
  $$
  \sigma = \sqrt{2.92} \approx 1.71
  $$

---

## 4. Example: Binomial Distribution

If $X \sim \text{Binomial}(n,p)$:

- **Mean:**
  $$
  \mathbb{E}[X] = np
  $$

- **Variance:**
  $$
  \text{Var}(X) = np(1-p)
  $$

---

## 5. Key Insights

- **Mean:** a measure of **center** (expected value).  
- **Variance:** a measure of **spread** (average squared distance from mean).  
- **Standard deviation:** the **spread in original units**, easier to interpret.  
- **General formulas** (sums) apply to all discrete random variables;  
  **closed-form results** exist for special distributions (binomial, Poisson, etc.).
