```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.power as smp
import statsmodels.stats.proportion as smprop
```

```python
x = np.array([168, 161, 167, 179, 184, 166, 198, 187, 191, 179])

print(x)

y = np.arange(11)
print(y)

np.mean(x)

np.median(x)

np.var(x, ddof=1)

np.sqrt(np.var(x, ddof=1))

# or:
np.std(x, ddof=1)

np.percentile(x, [0,25,50,75,100], method='averaged_inverted_cdf')

# or
np.percentile(x, list(range(0,110,10)), method='averaged_inverted_cdf')
```

```python
plt.hist(x)
plt.show()
```

```python
# A density histogram or empirical density of the heights
plt.hist(x, bins=8, edgecolor='black', color='red', alpha=0.7,
density=True)
plt.show()
```

```python
# Empirical cumulative distribution plot
plt.ecdf(x)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([168, 161, 167, 179, 184, 166, 198, 187, 191, 179])

fig, axs = plt.subplots(1, 2, figsize=(8, 4))

# with default whiskers
axs[0].boxplot([np.append(x, [235])])

# with whiskers covering 100% of the data
axs[1].boxplot([np.append(x, [235])], whis=(0, 100))

plt.show()



```
