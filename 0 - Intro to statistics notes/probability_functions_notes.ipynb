# Probability Functions: PMF, PDF, and CDF

---

## 1. PMF (Probability Mass Function)

- **Applies to:** Discrete random variables.  
- **Definition:**

$$
f(x) = P(X = x)
$$

- **Properties:**
  - $$ f(x) \geq 0 $$
  - $$ \sum_x f(x) = 1 $$

### Example (Fair Die ðŸŽ²)

Random variable \(X\) = outcome of one die throw.  
Possible values: \( \{1,2,3,4,5,6\} \)

$$
f(x) = \frac{1}{6}, \quad x = 1,2,3,4,5,6
$$

So for example:

$$
f(2) = P(X=2) = \frac{1}{6}
$$

## 2. PDF (Probability Density Function)

- **Applies to:** Continuous random variables.  
- **Definition:**

$$
P(a \leq X \leq b) = \int_a^b f(x) \, dx
$$

- **Key Point:**
  - $$ P(X=x) = 0 $$
  - Probabilities are given by **areas under the curve**.  
  - $$ f(x) \geq 0 $$
  - $$ \int_{-\infty}^{\infty} f(x) \, dx = 1 $$

### Example (Uniform Distribution on [0,1])

$$
f(x) = 
\begin{cases}
1, & 0 \leq x \leq 1 \\
0, & \text{otherwise}
\end{cases}
$$

Probability that \(X \in [0.2,0.5]\):

$$
P(0.2 \leq X \leq 0.5) = \int_{0.2}^{0.5} 1 \, dx = 0.3
$$

## 3. CDF (Cumulative Distribution Function)

- **Applies to:** Both discrete and continuous random variables.  
- **Definition:**

$$
F(x) = P(X \leq x)
$$

### Discrete Case (Die)

$$
F(2) = P(X \leq 2) = P(X=1) + P(X=2) = \frac{2}{6} = \frac{1}{3}
$$

### Continuous Case (Uniform [0,1])

$$
F(x) = \int_0^x 1 \, dt = x, \quad 0 \leq x \leq 1
$$

So:
- \(F(0.5) = 0.5\) â†’ 50% chance \(X \leq 0.5\)  
- \(F(1) = 1\) â†’ 100% chance \(X \leq 1\)
